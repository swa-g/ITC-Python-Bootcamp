## The Data

### Context

This database contains 76 attributes, but all published experiments refer to using a subset of 14 of them. In particular, the Cleveland database is the only one that has been used by ML researchers to this date. The "goal" field refers to the presence of heart disease in the patient. It is integer valued from 0 (no presence) to 4.

|ATTRIBUTES | DESCRIPTION |
|-----------|-------------|
|age:      |age in years|
|sex:       |(1 = male; 0 = female)|
|cp:        |chest pain type|
|trestbps:   |  resting blood pressure (in mm Hg on admission to the hospital)|
|cholserum:   |   cholestoral in mg/dl|
|fbs:      | (fasting blood sugar > 120 mg/dl)(1 = true; 0 = false)|
|restecg :   |   resting electrocardiographic results|
|thalach:    |   maximum heart rate achieved|
|exang:    |    exercise induced angina (1 = yes; 0 = no);  chest pain that occurs during physical activity|
|oldpeak:   |     ST depression induced by exercise relative to rest (noticeable decrease in the ST segment on an electrocardiogram (ECG) that occurs during physical exertion compared to when a person is at rest)|
|slope:   |    the slope of the peak exercise ST segment|
|ca:    |   number of major vessels (0-3) colored by flourosopy|
|thal:  |    3 = normal; 6 = fixed defect; 7 = reversable defect|
|target:  |     1 or 0|

#### Distribution Plots allow us to visualize the distribution of a data set. These plots are: 

- distplot
- jointplot
- pairplot
- rugplot
- kdeplot

#### Categorical Data Plots using seaborn to plot categorical data! There are a few main plot types for this:
- factorplot
- boxplot
- violinplot
- stripplot
- swarmplot
- barplot
- countplot



#### Matrix plots allow you to plot data as color-encoded matrices and can also be used to indicate clusters within the data.
e.g, seaborn's heatmap and clustermap

## Distribution Plots

### Resizing Seaborn Plots

```python
plt.figure(figsize=(12, 8))
sns.displot(df['age'])
```

```python
sns.displot(df['age'],kde=False)
```

```python
sns.displot(df['age'],kde=False,bins=40,color='red')
```

```python
sns.displot(df['age'],kde=False,bins=40)
# Notice matplotlib calls are still connected!
plt.xlim(50,60)
```

## Pairplots

Pairplots perform scatterplots and histograms for every single column in your data set. This means it could be a huge plot for large datasets! Use with caution, as it could take a long time for large datasets and the figures could be too small!

INFO: https://seaborn.pydata.org/generated/seaborn.pairplot.html

```python
iris = pd.read_csv('iris.csv')
```

```python
iris.head()
```

```python
sns.pairplot(iris)
```

```python
# Shows KDEs instead of histograms along the diagonal
sns.pairplot(iris, hue="species")
```
