## Python Operators

Python divides the operators into the following groups:

### Arithmetic Operators

| Operator | Name          | Example   |
|----------|---------------|-----------|
| +        | Addition      | x + y     |
| -        | Subtraction   | x - y     |
| *        | Multiplication| x * y     |
| /        | Division      | x / y     |
| %        | Modulus       | x % y     |
| **       | Exponentiation| x ** y    |
| //       | Floor division| x // y    |

### Assignment Operators

| Operator | Example   |
|----------|-----------|
| =        | x = 5     |
| +=       | x += 3    |
| -=       | x -= 3    |
| *=       | x *= 3    |
| /=       | x /= 3    |
| %=       | x %= 3    |
| //=      | x //= 3   |
| **=      | x **= 3   |
| &=       | x &= 3    |
| \|=      | x \|= 3   |
| ^=       | x ^= 3    |
| >>=      | x >>= 3   |
| <<=      | x <<= 3   |

### Comparison Operators

| Operator | Name                     | Example   |
|----------|--------------------------|-----------|
| ==       | Equal                    | x == y    |
| !=       | Not equal                | x != y    |
| >        | Greater than             | x > y     |
| <        | Less than                | x < y     |
| >=       | Greater than or equal to | x >= y    |
| <=       | Less than or equal to    | x <= y    |

### Logical Operators

| Operator | Example   |
|----------|-----------|
| and      | x and y   |
| or       | x or y    |
| not      | not x     |

### Identity Operators

| Operator | Example   |
|----------|-----------|
| is       | x is y    |
| is not   | x is not y|

### Membership Operators

| Operator | Example   |
|----------|-----------|
| in       | x in y    |
| not in   | x not in y|

### Bitwise Operators

| Operator | Name          | Example   |
|----------|---------------|-----------|
| &        | AND           | x & y     |
| \|       | OR            | x \| y    |
| ^        | XOR           | x ^ y     |
| ~        | NOT           | ~x        |
| <<       | Zero fill left shift | x << 2 |
| >>       | Signed right shift   | x >> 2 |

```python
# Arithmetic Operators
x = 10
y = 3
print("Addition:", x + y)         # 13
print("Subtraction:", x - y)      # 7
print("Multiplication:", x * y)   # 30
print("Division:", x / y)         # 3.333...
print("Modulus:", x % y)          # 1
print("Exponentiation:", x ** y)  # 1000
print("Floor division:", x // y)  # 3

# Assignment Operators
x = 5
x += 3
print("x += 3:", x)  # 8
x -= 3
print("x -= 3:", x)  # 5
x *= 3
print("x *= 3:", x)  # 15
x /= 3
print("x /= 3:", x)  # 5.0
x %= 3
print("x %= 3:", x)  # 2.0
x //= 3
print("x //= 3:", x)  # 0.0
x **= 3
print("x **= 3:", x)  # 0.0
x = 5
x &= 3
print("x &= 3:", x)  # 1
x |= 3
print("x |= 3:", x)  # 3
x ^= 3
print("x ^= 3:", x)  # 0
x = 5
x >>= 3
print("x >>= 3:", x)  # 0
x <<= 3
print("x <<= 3:", x)  # 40

# Comparison Operators
x = 5
y = 3
print("Equal:", x == y)              # False
print("Not equal:", x != y)          # True
print("Greater than:", x > y)        # True
print("Less than:", x < y)           # False
print("Greater than or equal to:", x >= y)  # True
print("Less than or equal to:", x <= y)     # False

# Logical Operators
x = True
y = False
print("Logical AND:", x and y)  # False
print("Logical OR:", x or y)    # True
print("Logical NOT:", not x)    # False

# Identity Operators
x = ["apple", "banana"]
y = ["apple", "banana"]
z = x
print("x is y:", x is y)        # False
print("x is z:", x is z)        # True
print("x is not y:", x is not y)  # True

# Membership Operators
x = ["apple", "banana"]
print("'banana' in x:", 'banana' in x)        # True
print("'cherry' not in x:", 'cherry' not in x)  # True

# Bitwise Operators
x = 5  # 0101 in binary
y = 3  # 0011 in binary
print("Bitwise AND:", x & y)   # 1 (0001 in binary)
print("Bitwise OR:", x | y)    # 7 (0111 in binary)
print("Bitwise XOR:", x ^ y)   # 6 (0110 in binary)
print("Bitwise NOT:", ~x)      # -6 (inverts all bits)
print("Zero fill left shift:", x << 2)  # 20 (10100 in binary)
print("Signed right shift:", x >> 2)    # 1 (0001 in binary)
```
